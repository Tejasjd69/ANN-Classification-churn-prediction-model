```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

import pickle
```

```python
df= pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
data=df.drop(["RowNumber","CustomerId","Surname"], axis=1)
```

```python
label_encode_gender=LabelEncoder()
data['Gender']=label_encode_gender.fit_transform(data['Gender'])
```

```python
##onehotencoder
from sklearn.preprocessing import OneHotEncoder
One_hot=OneHotEncoder()
geo_encode=One_hot.fit_transform(data[['Geography']])
geo_encode.toarray()
```

```python
with open('label_encode_gender.pkl' ,'wb') as file:
   pickle.dump(label_encode_gender,file) 

with open('geo_encode.pkl' ,'wb') as file:
   pickle.dump(One_hot,file) 
```

```python
## now separating dependent and independent variables

X=data.drop("Exited", axis =1) 
Y=data["Exited"]
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
scalar=StandardScaler()
X_train= scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
X_train
```

```python
with open("scalar.pkl","wb") as file:
  pickle.dump(scalar,file)
```

##ANN implementation


```python
import tensorflow as tf 

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping , TensorBoard
import datetime 


```

```python
model=Sequential([
 Dense(64,activation='relu',input_shape=(X_train.shape[1],)),   ##hideen layer 1
 Dense(32,activation='relu'),##hidden layer 2
 Dense(1,activation='sigmoid') ## output layer 




])
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss =tf.keras.losses.BinaryCrossentropy()   ## one way to it 
loss
```

```python
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
##lets create log table 

logs_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensor_callback=TensorBoard(log_dir=logs_dir, histogram_freq=1)
```

```python
Early_Stopping=EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
history=model.fit(
  X_train,Y_train,validation_data=(X_test,Y_test),epochs=100,
  callbacks=[tensor_callback,Early_Stopping]
)
```

```python
model.save('model.h5')
```
